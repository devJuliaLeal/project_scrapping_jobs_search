```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python


# Ler o arquivo CSV
df = pd.read_csv('data_all_pages.csv')

# Criar colunas vazias para as informações desejadas
df['tecnologias'] = None
df['nivel'] = None
df['remoto'] = None
df['salario'] = None

# Percorrer o DataFrame e extrair informações das colunas existentes
for index, row in df.iterrows():
    # Extrair as tecnologias da coluna 'tecnologia'
    tecnologias = row['tecnologia'].split(',')
    df.at[index, 'tecnologias'] = tecnologias

    # Extrair o nível do cargo com base na descrição do cargo
    descricao = row['nome']
    if 'sênior' in descricao.lower():
        nivel = 'Sênior'
    elif 'pleno' in descricao.lower():
        nivel = 'Pleno'
    elif 'júnior' in descricao.lower():
        nivel = 'Júnior'
    else:
        nivel = 'Não especificado'
    df.at[index, 'nivel'] = nivel

    # Verificar se a descrição do local de trabalho contém "Remoto"
    remoto = 'Remoto' if 'remoto' in row['outrosDetalhes'].lower() else 'Não Remoto'
    df.at[index, 'remoto'] = remoto
    
     
    # Extrair o valor do salário máximo da coluna 'outrosDetalhes'
    salario = row['outrosDetalhes'].split(',')[-2]
    salario = salario.split('R$')[-1].replace(',', '').strip()
    df.at[index, 'salario'] = salario

# Salvar o DataFrame atualizado em um novo arquivo CSV
df.to_csv('seu_arquivo_atualizado.csv', index=False)
```

```python



# Função para extrair o valor do salário máximo da coluna 'outrosDetalhes'
def extrair_salario_maximo(outros_detalhes):
    detalhes = outros_detalhes.split(',')
    salarios = []
    for detalhe in detalhes:
        if 'Até R$' in detalhe:
            salario = detalhe.split('Até R$')[-1].strip()
            salarios.append(salario)
    if salarios:
        salario_maximo = max(salarios)
        return salario_maximo
    else:
        return None

# Aplicar a função na coluna 'outrosDetalhes' e criar a nova coluna 'salario_maximo'
df['salario_maximo'] = df['outrosDetalhes'].apply(extrair_salario_maximo)

# Exibir o DataFrame atualizado
display(df)
```

```python

# Criar colunas de tecnologia
tecnologias = df['tecnologia'].str.split(',', expand=True)
num_colunas = tecnologias.shape[1]
for i in range(num_colunas):
    df[f'Tecnologia {i+1}'] = tecnologias[i]

# Exibir o DataFrame resultante
display(df)
```
